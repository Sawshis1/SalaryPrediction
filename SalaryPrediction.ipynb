```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("D:\AI\Salary.csv")
```

```python
dataset.head(5)
```

```python
plt.scatter(dataset['YearsExperience'],dataset['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=101)
```

```python
x_test
```

```python
x_train
```

```python
y_test
```

```python
y_train
```

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred_LR=LR.predict(x_test)
```

```python
y_pred_LR
```

```python
difference_LR=y_test-y_pred_LR
```

```python
res_df=pd.concat([pd.Series(y_pred_LR),pd.Series(y_test),pd.Series(difference_LR)],axis=1)
res_df.columns=['Prediction','Original Data','Difference']
```

```python
res_df
```

#Visualization

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,LR.predict(x_train),color='blue')
plt.title('Salary vs Exerience graph (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,LR.predict(x_train),color='blue')
plt.title('Salary vs Exerience graph (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

#Metrics

```python
from sklearn import metrics
rmse=np.sqrt(metrics.mean_squared_error(y_test,y_pred_LR))
R2=metrics.r2_score(y_test,y_pred_LR)
```

```python
rmse
```

```python
R2
```
